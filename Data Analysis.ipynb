```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
data=pd.read_csv(r"C:\Users\Subhada\Desktop\Data Analyst Assignment\Assignment file for Data Analyst - assignment file.csv")
```

```python
SOC=data.loc[:,"SOC"]
```

```python
Grid_Status=data.loc[:,"Grid status"]
```

```python
np.corrcoef(SOC,Grid_Status)
```

```python
plt.scatter(Grid_Status, SOC)
plt.xlabel("Grid_Status")
plt.ylabel("SOC")
plt.show()
```
